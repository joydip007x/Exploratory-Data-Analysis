```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
```

```python
data = pd.read_excel('drug_data_100_patients.xlsx')
```

 # female = 0, male = 1

```python
data['Sex'] = data['Sex'].replace(['F','M'],[0,1])
```

Blood pressure, HIGH = 2, LOW = 0, NORMAL = 1

```python
data['Blood Pressure'] =  data['Blood Pressure'].replace(['HIGH','LOW','NORMAL'],[2,0,1])
```

Cholesterol , HIGH = 1, NORMAL = 0

```python
data['Cholesterol'] = data['Cholesterol'].replace(['HIGH','NORMAL'],[1,0])
```

```python
data.head()
```

```python
age = data['Age']
mean_of_age = np.mean(age)
mean_of_age
```

```python
Ratio_of_Na_to_K =  data['Na_to_K Ration in Blood']
mean_of_Na_K = np.mean(Ratio_of_Na_to_K)
mean_of_Na_K
```

```python
median_of_age = np.median(age)
median_of_age
```

```python
median_of_Na_K = np.median(Ratio_of_Na_to_K)
median_of_Na_K
```

```python
mode_of_age = stats.mode(age)
mode_of_age
```

```python
mode_of_Blood_Pressure = stats.mode(data['Blood Pressure'])
mode_of_Blood_Pressure 
```

```python
mode_of_Cholesterol = stats.mode(data['Cholesterol'])
mode_of_Cholesterol
```
